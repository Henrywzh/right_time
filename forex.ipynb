```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
currencies = ['HKD=X', 'GBPHKD=X']
```

```python
df = yf.download(currencies, start='2010-01-01')
df['Close'].plot()
plt.xlabel('Date')
plt.ylabel('1USD, 1GBP: HKD')
plt.show()
```
