```python
import matplotlib.pyplot as plt

from finance_utils.backtest import Backtest
from finance_utils.utils import *
import yfinance as yf
import pandas as pd

start_date = '2000-01-01'
df = yf.download('3606.HK', start=start_date)

df_new = pd.DataFrame()
df_new['Price'] = df['Adj Close']
df_new['Value'] = df['Adj Close']
df_new['Return'] = df['Adj Close'].pct_change()

test_3606 = Backtest(df_new)
```

```python
plot_monthly_box(test_3606.results['Monthly Return'])
```

check +- 1 month price change for reports

```python
interim_dates = [
    '2015-08-31', '2016-09-05', '2017-08-10', '2018-08-23', '2019-09-11', '2020-09-03', 
    '2021-08-19', '2022-09-13', '2023-09-05', '2024-08-06'
]

interim_dates = pd.to_datetime(interim_dates)

for date in interim_dates:
    days = pd.to_timedelta(42, unit='d')
    data = df['Adj Close'].loc[date - days : date + days]
    fig, ax = plt.subplots(figsize=(10, 6))
    plt.plot(data, label='Price')
    plt.plot(date, data.loc[date], marker='^', color='r', label='Report Released')
    plt.legend()
    plt.ylabel('Price')
    plt.title('Impact of Interim Report Release to Stock Price')
    plt.show()
```
